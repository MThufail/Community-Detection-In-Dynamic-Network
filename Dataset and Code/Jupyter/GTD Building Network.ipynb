```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import networkx as nx
import math
```

```python
#Loading cleaned data 
df=pd.read_csv('data/globalterrorism_cleaned.csv',encoding='ISO-8859-1', low_memory=False)
print("Data imported successfully")
```

```python
# Build your graph
G=nx.from_pandas_edgelist(df, 'gname', 'country',create_using=nx.DiGraph())
# Plot it
nx.draw(G,node_color ='green')
plt.show()
print("\ngraph: {} nodes {} edges\n".format(len(G.nodes()), len(G.edges())))
```

```python
#Extra Code

g=nx.Graph()
for i in df.index[:]:
  target=df["country"][i]
  source=df["gname"][i]
  g.add_edge(*(source,target))
nx.draw(g, node_color ='red')
plt.show()

columns = list(df.columns.values)# Get columns name
g = nx.empty_graph(0, nx.DiGraph()) #initialize an empty graph
for i in range(len(columns)-1):
    g.add_edges_from(zip(df[columns[i]], df[columns[i+1]]))
    
# Plot it
nx.draw(g)
plt.show()
```
